### Blur

```python
def submatriz(matriz, i, j):
    sub = matriz[max(i-1,0):i+2,max(j-1,0):j+2]
    return sub
```

```python
def media_elementos(M):
    n = M.shape[0]*M.shape[1]
    soma_1 = np.sum(M, axis = 1)
    soma_2 = np.sum(soma_1, axis = 0)
    return soma_2/n
```

```python
def blur(foto):
    foto_borrada = foto.copy()
    M,N = foto.shape[0], foto.shape[1]
    for i in range(M):
        for j in range(N):
            foto_borrada[i][j] = media_elementos(submatriz(foto, i, j))
    return foto_borrada
```

```python
borrada = blur(img)
plt.imshow(borrada)
plt.axis("off")

plt.show()
```

Quando mais aplicamos a funo blur, mais borrada a imagem se torna:

```python
for i in range(10):
    if i == 0:
        borrada = blur(img)
    else:
        borrada = blur(borrada)
        
plt.imshow(borrada)
plt.axis("off")
plt.show()
```
